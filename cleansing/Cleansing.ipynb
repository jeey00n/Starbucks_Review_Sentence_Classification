```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
```

# 네이버

```python
naver = pd.read_csv('naver_store1_3.csv')
naver.head()
```

## 결측값 처리

```python
naver.isnull().sum()
```

```python
naver = naver.dropna()
naver
```

## DB를 위한 SiteID column(SiteDF의 PK) 생성 및 필요없는 column 제거

```python
naver['SiteID'] = 1
naver = naver.drop(['site', 'Unnamed: 0', 'user'], axis=1)
naver
```

```python
naver.to_csv('./naver0412.csv')
```

# Stores

```python
# store_info_DF
stores = pd.read_csv('starbucks_stores_info.csv')
stores
```

```python
reviewCount = naver.groupby('store').count().reset_index()[['store', 'review']]
reviewCount
```

```python
storesInfo = pd.merge(stores, reviewCount, on='store', how='left')
storesInfo
```

```python
storesInfo.to_csv('stores0412.csv')
```
